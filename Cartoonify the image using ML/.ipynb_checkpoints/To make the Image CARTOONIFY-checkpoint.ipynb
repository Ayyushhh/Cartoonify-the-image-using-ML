```python
import sys
```

### to open dialog box 

```python
imagepath = easygui.fileopenbox()
```

### to convert image into array and than resize it just to maintain the size all over the progrm

```python
originalimage = cv2.imread(imagepath)
originalimage = cv2.cvtColor(originalimage,cv2.COLOR_BGR2RGB)
    
if( originalimage is None):
    sys.exit()
        
resizeimage1 = cv2.resize(originalimage,(500,600))
plt.imshow(resizeimage1,cmap="gray")
```

### convert to gray scale image

```python
greyimage = cv2.cvtColor(originalimage,cv2.COLOR_BGR2GRAY)
resizegrayimage = cv2.resize(greyimage,(500,600))
plt.imshow(resizegrayimage,cmap="gray")
```

### smooth the image (blur )

```python
smoothimage = cv2.medianBlur(greyimage,7)
resizesmoothimage = cv2.resize(smoothimage,(500,600))
plt.imshow(resizesmoothimage,cmap="gray")
```

### take the edge of the image only

```python
edgeimage = cv2.adaptiveThreshold(smoothimage,255,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY,11,11) # greyiamge or smoothiamge
resizedgeimage = cv2.resize(edgeimage,(500,600))
plt.imshow(resizedgeimage,cmap="gray")
```

### mask the image or lighting the color 

```python
maskimage = cv2.bilateralFilter(originalimage,15,500,500)
resizemaskimage = cv2.resize(maskimage,(500,600))
plt.imshow(resizemaskimage,cmap="gray")
```

### mix the mask the edge image

```python
finalcartoonimage = cv2.bitwise_and(maskimage,maskimage,mask=edgeimage)
resizefinalimage = cv2.resize(finalcartoonimage,(500,600))
plt.imshow(resizefinalimage,cmap="gray")
```

### to show all images at onces

```python
images = [resizeimage1,resizegrayimage,resizesmoothimage,resizedgeimage,resizemaskimage,resizefinalimage]
fig, axes = plt.subplots(2,3, figsize=(12.,12), subplot_kw={'xticks':[], 'yticks':[]}, gridspec_kw=dict(hspace=0.1, wspace=0.1))
for i, ax in enumerate(axes.flat):
    ax.imshow(images[i], cmap='gray')

plt.show()
```
